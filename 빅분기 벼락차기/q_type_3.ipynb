```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/refs/heads/master/krdatacertificate/s1.csv')
display(df.head(10))
df.info()
```

```python
cls_1 = df[df['Classification'] == 1]['Resistin']
cls_2 = df[df['Classification'] == 2]['Resistin']
# print(cls_1)
print(cls_2)
```

```python
# f 검정 통계량 구하기

var1 = cls_1.var()
var2 = cls_2.var()

print(var1, var2)
print("---------------------------------")

dof_1 = len(cls_1) - 1
dof_2 = len(cls_2) - 1

print(dof_1, dof_2)
print("---------------------------------")

f_stat = var2/var1
print(round(f_stat, 3))
print("---------------------------------")
```

```python
# 합동 분산 추정량 구하기

pooled_var = (var1 * dof_1 + var2 * dof_2) / (dof_1 + dof_2)

print(round(pooled_var, 3))
print("---------------------------------")
```

```python
# t 검정 수행

import numpy as np
from scipy import stats

n1 = len(cls_1)
n2 = len(cls_2)

mean1 = cls_1.mean()
mean2 = cls_2.mean()

t_stat = (mean1 - mean2) / np.sqrt(pooled_var * (1/n1 + 1/n2))
p_value = 2*(1-stats.t.cdf(abs(t_stat), df=n1+n2-2))

print(round(p_value, 3))
print("---------------------------------")
```

```python
ttest_result = stats.ttest_ind(cls_1, cls_2, equal_var=True)
print(ttest_result)
print("---------------------------------")
```
