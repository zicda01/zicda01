````python
# 
# sequence     0   . 
#      '0 - len(sequence)'   .
# ,     3  step .

import datetime

df = df.rename(columns={ '(--:)': 'date'})

def func(date):
    date_str = str(date)
    if date_str[-2:] != '24':
        date_str = date_str.replace(':', ' ') #  replace     ,       .
        date = pd.to_datetime(date_str)
        
    else:
        ymd = date_str.split(':')[0]
        date = pd.to_datetime(ymd) + datetime.timedelta(days=1)
        
    return date


new_df = df['date'].map(func)
df['date'] = pd.to_datetime(new_df)
display(df)
````
