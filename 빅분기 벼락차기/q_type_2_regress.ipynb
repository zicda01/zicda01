````python
import pandas as pd

# Q. 학생의 성적 G3를 예측하라.

x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_test.csv")


display(x_train.head())
display(y_train.head())
````

````python
x_train.info()
y_train.info()
````

````python
# student id 컬럼을 제거한다.
x_train = x_train.drop(columns=['StudentID'])
x_test = x_test.drop(columns=['StudentID'])
y_train = y_train.drop(columns=['StudentID'])
````

````python
print(x_train)
print(y_train)
print(x_test)
````

````python
# 훈련 데이터(x_train)만 사용하여 통계량을 전달하기 위한 시리얼 생성.
number_columns = x_train.select_dtypes(include=['number']).columns.to_list()
category_columns = x_train.select_dtypes(exclude=['number']).columns.to_list()

# 수치형 자료에 대해서는 중앙값인 median를 사용
train_median = x_train[number_columns].median()
# 범주형 자료에 대해서는 최빈값인 mode를 사용
train_mode = x_train[category_columns].mode().iloc[0]

print(train_median)
print(train_mode)

# 평균값(mean)은 이상치에 민감하기 때문에, median과 mode를 사용하는 것이 보편적이다.
````

````python
# 학습 데이터의 통계량을 이용해서 모든 데이터셋의 결측치를 대치함.
x_train[number_columns] = x_train[number_columns].fillna(train_median)
x_train[category_columns] = x_train[category_columns].fillna(train_mode)

# 테스트 데이터에도 동일한 작업을 수행해준다.
x_test[number_columns] = x_test[number_columns].fillna(train_median)
x_test[category_columns] = x_test[category_columns].fillna(train_mode)
````

````python
# 수치형 데이터에 대해서 스케일링을 수행한다.
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

x_train[number_columns] = scaler.fit_transform(x_train[number_columns])
x_test[number_columns] = scaler.transform(x_test[number_columns])
````

````python
# 범주형 데이터에 대해서 라벨 인코딩을 수행한다.
from sklearn.preprocessing import LabelEncoder

# 각 범주형 컬럼에 대해 반복문으로 LabelEncoder 적용, 컬럼을 하나씩 인코딩합니다.
for col in category_columns:
    # LabelEncoder 객체 생성
    ## 인코더 객체가 한 번에 하나의 컬럼만 처리하도록 설계되어 있으므로, 초기화 작업이 필요합니다.
    encoder = LabelEncoder()
    
    # X_train을 기준으로 인코딩 기준(클래스)을 정합니다.
    x_train[col] = encoder.fit_transform(x_train[col])
    
    # x_train에서 학습된 기준으로 x_test를 변환합니다.
    x_test[col] = encoder.transform(x_test[col])
````

````python
# 학습/검증 데이터를 분리합니다.
from sklearn.model_selection import train_test_split
X_train, X_validate, Y_train, Y_validate = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
````

````python
# 머신러닝 학습을 수행하고, 학습된 모델을 생성합니다.
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, Y_train)
y_val_pred = rfr.predict(X_validate)

# print(y_val_pred)
````

````python
# 학습된 모델의 성능을 검증 데이터를 통해 테스트합니다.
from sklearn.metrics import root_mean_squared_error, r2_score
rmse = root_mean_squared_error(Y_validate, y_val_pred)
r2 = r2_score(Y_validate, y_val_pred)

print(rmse)
print(r2)
````

````python
y_pred = rfr.predict(x_test)
# print(y_pred)
result = pd.DataFrame(y_pred, columns=['pred'])
result.to_csv('result_csv', index=False)
````
