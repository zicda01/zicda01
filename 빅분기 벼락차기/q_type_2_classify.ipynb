```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")


display(x_train.head())
display(y_train.head())
```

```python
# 문제. 고객의 신상정보 데이터를 통한 회사 서비스 이탈 예측 (종속변수 : Exited)
```

```python
x_train.info()
print("-----------------------------------------------")
y_train.info()
```

1. 단순 결측치는 없는 것으로 확인된다.
2. 학습, 검증을 위한 데이터셋 x 와 예측하고자 하는 종속변수 y 가 분리되어 제공되고 있다.

```python
# x_test 데이터를 확인

display(x_test)
x_test.info()
```

```python
# concat() 을 수행하여 학습을 위한 데이터를 만든다.
## pd.concat() : 두 개의 데이터 프레임을 id에 맞춰서 결합시킨다.
### 이 작업은 학습을 위해서, 독립변수와 종속변수를 하나의 데이터 프레임으로 만들기 위해 필요하다.

train_set = pd.concat([x_train, y_train['Exited']], axis = 1)

display(train_set.head(20))
train_set.info()
```

```python
# describe()를 이용해 전처리를 수행하기 이전, 데이터 정보를 파악할 수 있다.
## 1. 범주형 자료의 정보를 파악한다.
display(train_set.describe(include=['object', 'category']))
```

범주형 자료에 대한 .describe() 메서드 수행 결과 변수
   - count: 결측치를 제외한 전체 데이터의 수 (행의 수와 일치).
   - unique: 서로 다른 범주의 수 (범주 종류의 수).
   - top: 최빈 범주
   - freq: 최빈 범주의 개수.

```python
## 2. 수치형 자료의 정보를 파악한다.
### describe()메서드에 인자를 따로 전달하지 않으면, 기본적으로 수치형 자료에 대해서 출력된다.

display(train_set.describe())
```

```python
# 과적합을 방지하기 위해, 사용자의 고유한 id 컬럼을 제거한다.
## surname 컬럼 또한, 예측 데이터로서 무의미. unique value 가 매우 많다고 나오기 때문에(상식적으로도 그렇지만, 기술적으로 그렇게 해석할 수 있다.)

train_set = train_set.drop(columns=['CustomerId', 'Surname'])
display(train_set)
```

```python
# 검증용 데이터에도 동일한 작업을 수행해준다.

x_test = x_test.drop(columns=['CustomerId', 'Surname'])
display(x_test)
```

```python
x = train_set.drop(columns=['Exited'])
y = train_set['Exited']

display(x)
display(y)
```

```python
# 범주형 데이터를 학습할 수 있도록 encoding 작업이 필요함.
## 수치형 데이터를 전달하기 위해서는 'number' 키워드를 사용해야 함. 'numeric' 은 오류 발생.

category_lst = train_set.drop(columns=['Exited']).select_dtypes(exclude=['number']).columns.to_list()
print(category_lst)

x_encoded = pd.get_dummies(x, columns=category_lst, drop_first=False)
x_test_encoded = pd.get_dummies(x_test, columns=category_lst, drop_first=False)

# display(x_encoded)
```

원-핫 인코딩 적용
- drop_first = False
- False: 기준 범주 없이 모든 범주를 0과 1로 encoding 할 때 사용한다. 이는 기준 범주 없이 개별 범주로 처리하기 때문.
- True: 통계모델에서 기준 범주를 설정할 때 사용한다. 통계에서는 기준 범주를 기준으로하여 변화량을 측정하기 때문. (이 경우 기준범주가 0이 된다).

<hr>

- 위 사항은 모두 이론적인 부분이니 당장은 그냥 넘어가도 무방. 빅분기 시험에 대해선 drop_first=False 를 인자로 전달하는 것이 유리하다고만 기억하자. (애초에 머신러닝은 보편적으로 그렇게 하는 게 맞다고 함 by gemini)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_validate, y_train, y_validate = train_test_split(x_encoded, y, test_size=0.3, random_state=42, stratify=y)
```

학습용 데이터와 검증용 데이터를 분리하는 작업.
특히 **train_test_split()** 메서드에 **stratify=y** 를 전달하여, 예측값의 분포 비율을 일치시키는 게 중요하다.

```python
# 수치형 데이터에 대한 스케일링 작업.
## 스케일링은 수치형 데이터에 대해서, 데이터의 기술 통계량을 데이터에 반영시켜서 머신러닝 성능을 좀 더 안정적으로 만들 수 있다.

number_lst = X_train.select_dtypes(include=['number']).columns.to_list()
# print(number_lst)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# 학습용 데이터에는 scaler.fit_transform() 메서드를 수행한다.
X_train[number_lst] = scaler.fit_transform(X_train[number_lst])

# 검증용 데이터에는 scaler.transform() 메서드를 수행한다.
X_validate[number_lst] = scaler.transform(X_validate[number_lst])

# 최종 테스트 데이터에 대해서도 스케일링을 해준다. (데이터 처리 로직의 일관성 유지)
x_test_encoded[number_lst] = scaler.transform(x_test_encoded[number_lst])
```

스케일링은 단순히 값을 바꾸는 것이 아니라, 각 특징(Feature)이 가지는 범위와 크기의 격차를 해소하여, "모델이 특정 큰 값에만 치우쳐 학습하는 것을 막는" 공정성을 확보하는 과정이라고 이해하시면 됩니다. <br>
\- by gemini

```python
# 실제로 랜덤포레스트 학습을 실시하여, 머신러닝 모델을 만든다.

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=23)

rfc.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix

# 1. 검증 데이터에 대한 예측
# 클래스 예측 (0 또는 1)
val_pred_label = rfc.predict(X_validate)
# 양성 클래스(1)일 확률 예측 (AUC 계산에 사용)
val_pred_proba = rfc.predict_proba(X_validate)[:, 1]

# 2. AUC-ROC 지표 출력 (빅분기 2유형 주요 채점 지표)
auc_score = roc_auc_score(y_validate, val_pred_proba)
print(f"✅ 검증 데이터 AUC-ROC: {auc_score:.4f}")

# 3. 상세 평가 지표 출력
print("\n📋 Classification Report:")
print(classification_report(y_validate, val_pred_label))

print("🧮 Confusion Matrix:")
print(confusion_matrix(y_validate, val_pred_label))
```

학습 데이터와 검증 데이터를 이용해서, 모델을 평가하는 작업. 그러나 나는 그냥 낼 것임...<br>
실제 모델을 개선하는 작업을 수행한다면, 끊임 없이 검증 데이터를 확인하고, 성능을 올려야 함.

```python
# 최종 예측 데이터를 생성합니다.
test_pred_label = rfc.predict(x_test_encoded)
test_pred_proba = rfc.predict_proba(x_test_encoded)[:, 1]

# 최종 예측 데이터를 CSV로 저장합니다.
pd.DataFrame({'pred': test_pred_label}).to_csv('label.csv', index=True, index_label='ID')
pd.DataFrame({'pred': test_pred_proba}).to_csv('proba.csv', index=True, index_label='ID')
```
